```python
# Dependencies and set-up
import pandas as pd
import requests
import json
from config import api_key
from pprint import pprint
from datetime import date
from matplotlib import pyplot as plt
import numpy as np
import gmaps
from config import g_key
import scipy.stats as stats
from datetime import date
```

```python
# Base url for Covid-19 statistics API
url = "https://covid-19-coronavirus-statistics.p.rapidapi.com/v1/stats"

# Query parameters
querystring = {"country":"US"}

headers = {
    'x-rapidapi-host': "covid-19-coronavirus-statistics.p.rapidapi.com",
    'x-rapidapi-key': api_key
    }


# Attempt to retrieve county level data for the United States
try:
    response = requests.get(url, headers=headers, params=querystring).json()
    
# Print error message if error occurs during API call
except:
    print('API call failed.')

# Pretty print the response
pprint(response)

```

```python
# List of county data from every US county
county_data =[county for county in response['data']['covid19Stats']]

# Data frame of statistics for each county in the United States
usa_df = pd.DataFrame(county_data)

# Clean up the column names
renamed_usa_df = usa_df.rename(columns={
    'city': 'County',
    'province': 'State',
    'country': 'Country',
    'lastUpdate': 'Date',
    'keyId': 'Key ID',
    'confirmed': 'Confirmed Cases',
    'deaths': 'Deaths',
    'recovered': 'Recovered'
})

# View the cleaned up data frame
renamed_usa_df
```

```python
# Replace empty string values with NaN
renamed_usa_df.replace('', np.nan, inplace=True)
renamed_usa_df
```

```python
# Drop rows with NaN values
dropped_usa_df = renamed_usa_df.dropna()
dropped_usa_df
```

```python
# View a list of unique state names
unique_states = dropped_usa_df['State'].unique()
unique_states
```

```python
# Create a list of the unique state names
unique_states_list = list(unique_states)
```

```python
# Remove any rows where the state is 'Northern Mariana Islands' of 'District of Columbia'
unique_states_list.remove('Northern Mariana Islands')
unique_states_list.remove('District of Columbia')
```

```python
# View the list of states in alphabetical order
fifty_states = sorted(unique_states_list)
fifty_states
```

```python
# Import U.S. Census population data for the United States from a .csv
population_data = pd.read_csv('Resources/county_population_data.csv', encoding='iso-8859-1')
# Dataframe of the .csv
population_df = pd.DataFrame(population_data)
# Dataframe of just county and 2019 population estimate columns
county_pop_df = population_df[['CTYNAME', 'POPESTIMATE2019']]
# Rename columns
county_pop_df.columns = ['County', 'Population']
county_pop_df
```

```python
# Create dataframe from the list of U.S. states 
fifty_states_df = pd.DataFrame(fifty_states)
# Rename column
fifty_states_df.columns = ['County']
# Combine the state names and populations into a single data frame
state_pop_df = fifty_states_df.merge(county_pop_df, on='County', how='left')
# Rename columns
state_pop_df.columns = ['State', 'Population']
state_pop_df
```

```python
# Group the Covid-19 USA dataframe by state
grouped_states = renamed_usa_df.groupby('State')
# Sum the total values for for each state and reset the index
state_totals = grouped_states.sum().reset_index()
# View the dataframe
state_totals
```

```python
# Merge the Covid-19 data with the state population data
covid_pop_df = state_pop_df.merge(state_totals, on='State', how='left')
# Delete the 'Recovered' column
del covid_pop_df['Recovered']
# Export the dataframe to a .csv
covid_pop_df.to_csv('Resources/covid_pop.csv', index=False)
covid_pop_df
```

```python
# Add columns for cases and deaths per 100,000 residents
covid_pop_df['Cases Per 100,000'] = covid_pop_df['Confirmed Cases']/covid_pop_df['Population']*100000
covid_pop_df['Deaths Per 100,000'] = covid_pop_df['Deaths']/covid_pop_df['Population']*100000
covid_pop_df['Death Rate'] = covid_pop_df['Deaths']/covid_pop_df['Confirmed Cases']
covid_pop_df
```

```python
# Sort the dataframe and store the results

#Sort dataframe by number of cases
by_case_num = covid_pop_df.sort_values('Confirmed Cases', ascending=False).reset_index(drop=True)
#Sort dataframe by cases per capita
by_case_per = covid_pop_df.sort_values('Cases Per 100,000', ascending=False).reset_index(drop=True)
#Sort dataframe by number of deaths
by_death_num = covid_pop_df.sort_values('Deaths', ascending=False).reset_index(drop=True)
#Sort dataframe by deaths per capita
by_death_per = covid_pop_df.sort_values('Deaths Per 100,000', ascending=False).reset_index(drop=True)
```

```python
# Store the top ten states sorted for each value

# Top 10 states by number of cases
top_cases = by_case_num.head(10)

# Top 10 states by cases per 100,000 residents
top_cases_per = by_case_per.head(10)

# Top 10 states by number of deaths
top_deaths = by_death_num.head(10)

# Top 10 States by deaths per 100,000 residents
top_deaths_per = by_death_per.head(10)
```

```python
# Plot the number of deaths per 100,000 residents for each of the top ten states based on total death number
plt.bar(top_deaths['State'], top_deaths['Deaths Per 100,000'])

# Format x-tick label
plt.xticks(ha='right', rotation=45)

# Axis labels
plt.xlabel('Top Ten States (by total number of deaths)')
plt.ylabel('Number of Deaths (per 100,000 residents)')
plt.title('''
Number of Deaths per 100,000 Residents 
for the Top Ten States
Ranked by Total Number of Deaths
''')

# Export plot as .png
plt.savefig('Resources/top_deaths.png', bbox_inches='tight')

plt.show()
```

```python
# Plot the number of cases per 100,000 for each of the top ten states based on total case case number
plt.bar(top_cases['State'], top_cases['Cases Per 100,000'])

# Format x-tick label
plt.xticks(ha='right', rotation=45)

# Axis labels
plt.xlabel('Top Ten States (by total number of cases)')
plt.ylabel('Number of Cases (per 100,000 residents)')
plt.title('''
Number of Cases per 100,000 Residents 
for the Top Ten States
Ranked by Total Number of Casess
''')
# Export plot as .png
plt.savefig('Resources/top_cases.png', bbox_inches='tight')

plt.show()
```

```python
state_coords = pd.read_csv('../../state_coordinates.csv')
state_coords.columns = ['Abbr', 'Lat', 'Long', 'State']
stats_locations = covid_pop_df.merge(state_coords, on='State', how='left')
stats_locations

#Configure gmaps with api key
gmaps.configure(api_key=g_key)

#Store the lat/long values for each state
locations = stats_locations[['Lat', 'Long']].astype(float)

#Store the death per 100,000 values for each state
deaths_per_capita = stats_locations['Deaths Per 100,000'].astype(float)


# Create a heatmap of deaths per 100,000 for each state
fig = gmaps.figure()

#Create the heat map layer
death_layer = gmaps.heatmap_layer(locations, weights=deaths_per_capita, 
                                 dissipating=False, max_intensity=deaths_per_capita.max(),
                                 point_radius = 2)
#Add the heat map layer to the figure
fig.add_layer(death_layer)

fig

```

```python
# Add the state markers to the heatmap
info_box_template = """
<dl>
<dt>Deaths Per 100,000</dt><dd>{Deaths Per 100,000}</dd>
<dt>State</dt><dd>{State}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
state_info = [info_box_template.format(**row) for index, row in stats_locations.iterrows()]

# Create symbol layer
symbol_layer = gmaps.marker_layer(
    locations,
    info_box_content=state_info)

#Create a combined map of the heat map and hotel symbol layers
fig = gmaps.figure()

#Add the heat map and hotel symbol layers

fig.add_layer(symbol_layer)
fig.add_layer(death_layer)
#Display the combined map
fig
```

```python
# Dataframe of statues sorted by death rate
by_death_rate = covid_pop_df.sort_values(by='Death Rate', ascending=False)

# Store the total us population count
total_population = by_death_rate['Population'].sum()
total_population
# Store Californias population count
cali_population = by_death_rate.loc[by_death_rate['State'] == 'California', 'Population'].sum()
cali_population
```

```python
time_series = pd.read_csv('Resources/time_series_covid19_confirmed_us.csv')
time_series_df = pd.DataFrame(time_series)
time_series_group = time_series_df.groupby('Province_State').sum()
# Remove un-needed columns
del time_series_group['UID']
del time_series_group['code3']
del time_series_group['FIPS']
del time_series_group['Lat']
del time_series_group['Long_']

# Transpose dataframe rows and columns
pivot_group = time_series_group.T

# Sum the cases of all states for each day
column_sums = time_series_group.sum()
# Dataframe of summed columns
national_df = pd.DataFrame(column_sums)
cali = pivot_group['California'].copy()
cali_df = pd.DataFrame(cali)

cali_diff = cali_df.diff()
cali_diff_df = pd.DataFrame(cali_diff)
cali_change = cali_df.pct_change()
cali_change_df = pd.DataFrame(cali_change)
cali_diff_df
```
