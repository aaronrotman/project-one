```python
#Dependencies and set-up
import pandas as pd
import requests
import json
from config import api_key
from pprint import pprint
from datetime import date
from matplotlib import pyplot as plt
import numpy as np
```

```python
#Base url for Covid-19 statistics API
url = "https://covid-19-coronavirus-statistics.p.rapidapi.com/v1/stats"

#Query parameters
querystring = {"country":"US"}

headers = {
    'x-rapidapi-host': "covid-19-coronavirus-statistics.p.rapidapi.com",
    'x-rapidapi-key': api_key
    }


#Attempt to retrieve county level data for the United States
try:
    response = requests.get(url, headers=headers, params=querystring).json()
    
#Print error message if error occurs during API call
except:
    print('API call failed.')

#Pretty print the response
pprint(response)

```

```python
#List of county data from every US county
county_data =[county for county in response['data']['covid19Stats']]

#Data frame of statistics for each county in the United States
usa_df = pd.DataFrame(county_data)

#Clean up the column names
renamed_usa_df = usa_df.rename(columns={
    'city': 'County',
    'province': 'State',
    'country': 'Country',
    'lastUpdate': 'Date',
    'keyId': 'Key ID',
    'confirmed': 'Confirmed Cases',
    'deaths': 'Deaths',
    'recovered': 'Recovered'
})

#View the cleaned up data frame
renamed_usa_df
```

```python
#Replace empty string values with NaN
renamed_usa_df.replace('', np.nan, inplace=True)
renamed_usa_df
```
